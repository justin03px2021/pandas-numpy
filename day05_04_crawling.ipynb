```python
from bs4 import BeautifulSoup
import requests as req
```

```python
#긁어오기 싶은 페이지 지정
# 200은 성공 404는 페이지 없음

url = 'https://www.naver.com/'

#페이지 요청
response = req.get(url)
```

```python
#응답받은 내용 확인
response.text
```

```python
#우리가 원하는 글을 긁어오기
# id 는 # class 는 .
# 비정형 텍스트를 반정형으로 바꾸기
html = BeautifulSoup(response.text, 'html.parser')
html
```

```python
html.select('#account > p')[0].text
```
