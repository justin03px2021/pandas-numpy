```python
from bs4 import BeautifulSoup
import requests as req
```

```python
#긁어오고 싶은 사이트
url = 'https://apache.org/'
```

```python
#페이지 요청
response = req.get(url)
```

```python
#비정형을 반정형으로 변환
response.text
html = BeautifulSoup(response.text)
```

```python
# community led development apache way 불러오기
html.select('#main-header > div.main > h2')[0].text
```

```python
#apache 에서 관리하고 있는 프로젝트 리스트 가져오기
row = html.select('#projects-list > div > div > div.col-sm-10 > div')
```

```python
row[0].findAll('a')[4].text
```

```python
# a태그에 감싸진 모든 프로젝트 이름을 조회해주세요
for a in row[0].findAll('a'):
    print(a.text)
```
